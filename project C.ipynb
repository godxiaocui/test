```python
from sklearn.cluster import KMeans 
from sklearn import preprocessing
import pandas as pd
import numpy as np
```

```python
#写入数据
data=pd.read_csv('D:\data analayse\python\last testing\ProjectC\CarPrice_Assignment.csv')
train_x=data.drop(['CarName','car_ID'],axis=1)
train_x
```

```python
#将部分值改成数字，调用labelencoder
from sklearn.preprocessing import LabelEncoder
le =LabelEncoder()
#批量改数据
train_x['fueltype']=le.fit_transform(train_x['fueltype'])
train_x['aspiration']=le.fit_transform(train_x['aspiration'])
train_x['doornumber']=le.fit_transform(train_x['doornumber'])
train_x['carbody']=le.fit_transform(train_x['carbody'])
train_x['drivewheel']=le.fit_transform(train_x['drivewheel'])
train_x['enginelocation']=le.fit_transform(train_x['enginelocation'])
train_x['enginetype']=le.fit_transform(train_x['enginetype'])
train_x['cylindernumber']=le.fit_transform(train_x['cylindernumber'])
train_x['fuelsystem']=le.fit_transform(train_x['fuelsystem'])
train_x.head()
```

```python
#规范数据到0.1之间
min_mx_scaler=preprocessing.MinMaxScaler()
train_x=min_mx_scaler.fit_transform(train_x)
pd.DataFrame(train_x).to_csv('temp.csv',index=False)
train_x
```

```python
#手肘法算K的取值范围
import matplotlib.pyplot as plt
sse = []
for k in range(1, 100):
	# kmeans算法
	kmeans = KMeans(n_clusters=k)
	kmeans.fit(train_x)
	# 计算inertia簇内误差平方和
	sse.append(kmeans.inertia_)
x = range(1, 50)
plt.xlabel('K')
plt.ylabel('SSE')
plt.plot(x, sse, 'o-')
plt.show()
print(sse)
#k= 7 为最佳值
```

```python
# Kmeans 聚类
kmeans=KMeans(n_clusters=7)
#训练
kmeans.fit(train_x)
#预测结果
predict_x=kmeans.predict(train_x)
print(predict_x)
print(train_x)
```

```python
#合并聚类结果，插入到原数据
data['labels']=predict_x
result=data.sort_values("labels",ascending=False)
result.head()
result.to_csv("D:\\data analayse\\python\\last testing\\ProjectC\\car_data121.csv",encoding="gbk")
```
